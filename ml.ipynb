```python
y=df['Outcome']
x=df.drop('Outcome',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression()
dr = DecisionTreeClassifier()
```

```python
lr.fit(x_train,y_train)
dr.fit(x_train,y_train)
```

```python
print(lr.score(x_train,y_train))
```

```python
print("logistic",lr.score(x_train,y_train),lr.score(x_test,y_test))
```

```python
print("decision tree",dr.score(x_train,y_train),dr.score(x_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lr,x_test,y_test,cv=5)
```

```python
#Setup GridSearch with cross-val 
grid_search = GridSearchCV(estimator = lr,param_grid=param_grid,cv=5,scoring='accuracy')
```

```python
#fit the model with the training data
grid_search.fit(x_train,y_train)
```

```python
#best hyperparametry found
best_params = grid_search.best_params_
print("Best Parameters:",best_params)
```
